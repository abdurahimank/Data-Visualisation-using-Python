# Format Strings
You can use also use the shortcut string notation parameter to specify the marker.

This parameter is also called **fmt**, and is written with this syntax:

"**marker|line|color**"

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])

plt.plot(y, 'o:r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])

plt.plot(y, '-.r')  # without marker
plt.show()
```

### Using only markers

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3, 8, 1, 10])

plt.plot(y, '*r', ms=20)  # without line
plt.show()
```
